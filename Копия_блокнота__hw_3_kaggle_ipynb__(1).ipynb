```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
data_origin=data.copy()
```

```python
data_origin
```

```python
#  
St_scale = StandardScaler()
data[num_cols] = St_scale.fit_transform(data[num_cols])
data[num_cols]
```

```python
#   LabelEncoder    
encoders = {}
for _cols in cat_cols:
    lb_encoder = LabelEncoder()
    data[_cols] = lb_encoder.fit_transform(data[_cols])
    encoders[_cols] = lb_encoder
data[cat_cols]
```

```python
#    Onehotencoding
onhot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

train_encoded = onhot_encoder.fit_transform(data[cat_cols])
train_encoded_df = pd.DataFrame(train_encoded, columns=onhot_encoder.get_feature_names_out(onhot_encoder.feature_names_in_))
df = pd.concat([data, train_encoded_df], axis=1)
```

```python
df
```

```python
data_origin
```

```python
# 
X = df.drop([target_col],axis=1)
X_origin = data_origin.drop([target_col], axis=1)
y = df[target_col]
X_train,X_test,y_train,y_test = train_test_split(X.values, y.values, test_size = 0.3, random_state=42, stratify=y)
```

```python
X_train.shape, X_test.shape
```

```python
#    (  GridSearchCV)
params = {
    'C': [100, 10, 1, 0.1, 0.01, 0.001],
    'penalty': ['l2', 'l1'],
    'solver': ['liblinear', 'saga']
}
# 
model = LogisticRegression(random_state=42)
#  GridSearchCV
gs = GridSearchCV(
    model,
    params,
    cv=5,
    verbose=10,
    scoring='roc_auc',
    refit=True
)
#
gs.fit(X_train,y_train)

best_model = gs.best_estimator_

print(' : ', gs.best_params_)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
y_pred = best_model.predict(X_test)
print('accuracy: ', accuracy_score(y_test, y_pred))
print('precision: ', precision_score(y_test, y_pred))
print('recall: ', recall_score(y_test, y_pred))
print('f1: ', f1_score(y_test, y_pred))
print('ROC-AUC: ', roc_auc_score(y_test, y_pred))
```

```python
import catboost
```

```python
X_train_origin, X_test_origin, _, _ = train_test_split(X_origin.values, y.values,
                                                       test_size=0.3,
                                                       random_state=42)
```

```python
boost_model = catboost.CatBoostClassifier(n_estimators=400, cat_features=list(range(3, len(X_origin.columns))))

boost_model.fit(X_train_origin, y_train, cat_features=list(range(3, len(X_origin.columns))))

y_train_predicted = boost_model.predict_proba(X_train_origin)[:, 1]
y_test_predicted = boost_model.predict_proba(X_test_origin)[:, 1]
```

```python
train_auc = roc_auc_score(y_train, y_train_predicted)
test_auc = roc_auc_score(y_test, y_test_predicted)

print(train_auc, test_auc)
```

```python
#   CatBoost
params_cat = {
    'n_estimators': [50,100,200],
    'depth': [4, 6, 8],
    'learning_rate': [0.01, 0.03, 0.1]
}
model_cat = catboost.CatBoostClassifier(cat_features=list(range(3, len(X_origin.columns))))
#  GridSearchCV
gs_cat = GridSearchCV(
    model_cat,
    params_cat,
    cv=5,
    verbose=1,
    n_jobs=-1,
    scoring='roc_auc' ,
    refit=True
)
#
gs_cat.fit(X_train_origin,y_train)

best_model_cat = gs_cat.best_estimator_

print(' : ', gs_cat.best_params_)
```

```python
y_train_predicted_gs = best_model_cat.predict_proba(X_train_origin)[:, 1]
y_test_predicted_gs = best_model_cat.predict_proba(X_test_origin)[:, 1]
```

```python
train_auc_gs = roc_auc_score(y_train, y_train_predicted_gs)
test_auc_gs = roc_auc_score(y_test, y_test_predicted_gs)

print(train_auc_gs, test_auc_gs)
```
